```python
# q2.whats the avg unit price by country top 5
```

```python
df.groupby("Country")["Unit Price ($)"].mean().head(5)
```

```python
# q3. what is the total Amount($) by country top 5
```

```python
df.groupby("Country")["Amount($)"].sum().sort_values(ascending=False).head(5)
```
